```python
# psycopg2 acting as a bridge that connecting jupyter notebook with postgres

import psycopg2
```

# Create a connection to Database

```python
# The below syntax is used for connecting to database present in the sql server.

try:
    conn=psycopg2.connect(host='localhost', user='postgres',dbname='postgres', password='password')
except psycopg2.Error as e:
    print("Error: Could not make connection to the postgres database")
    print(e)
```

Use this connnection to get a cursor that can be used to execute queries.

```python
try:
    cur=conn.cursor()
except psycopg2.Error as e:
    print("Error: could not get curser to the database")
    print(e)
    
```

set automatic commit to be true so that each action is commited without having to call conn.commit() after each command.

```python
conn.set_session(autocommit=True)
```

# Create a database in the PostgreSQL environment

```python

# Create database ppdatabase nothing but python postgre database

try:
    cur.execute('create database ppdatabase')
except psycopg2.Error as e:
    print('Error: could not get cursor to database')
    print(e)
    
# The same thing can be seen postgres shell.
```

```python
# creating connection
try:
    conn=psycopg2.connect(host='localhost',user='postgres',dbname='ppdatabase',password='password')
except psycopg2.Error as e:
    print("Error: Connection to the database could not be done")
    print(e)
```

```python
# creating cursor

try:
    cur=conn.cursor()
except psycopg2.Error as e:
    print("Error: Couldn't create the cursor")
    print(e)
```

```python
conn.set_session(autocommit=True)
```

# Create a table in the database

Create Sales table for different products with their prices and quantity.

Product_id

availability

quantity

price


```python
# This is simple sql Qurie used for creating tables in database but there are now values.
try:
    cur.execute("create table if not exists sales (product_id int,availability varchar,quantity int,price float);")
except psycopg2.Error as e:
    print("Can not create a table in the database")
    print(e)
```

```python
# This is the sql qurie used for inserting values in to the table.

try:
    cur.execute("insert into sales (product_id,availability,quantity,price )\
                values(%s,%s,%s,%s)",
                (10021,'available',10,340.45)
                )
except psycopg2.Error as e:
                print("Error: Values couldn't be inserted")
                print(e)
                
```

```python
try:
    cur.execute("insert into sales (product_id,availability,quantity,price )\
                values(%s,%s,%s,%s)",
                (10022,'not available',4,40.45)
                )
except psycopg2.Error as e:
                print("Error: Values couldn't be inserted")
                print(e)
```

```python
# Retrive Data from the table sales.

try:
    cur.execute('select * from sales')
except Error.psycopg2 as e:
    print("Error: Table couldn't retrived")
    print(e)
```

```python
# now print each row in the table

row=cur.fetchone()

while row:
    print(row)
    row=cur.fetchone()
    
```

```python
# close cursor and connection.
cur.close()
conn.close()
```
